# Exploratory Data Analysis (EDA)

## Objective
To understand the structure, distribution, correlation, and characteristics 
of the Breast Cancer dataset before applying dimensionality reduction 
and classification models.

EDA will help justify:
- Need for scaling
- Need for dimensionality reduction
- Feature redundancy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\ML PROJECT\Dimensionality-Reduction-Classification-Study\data\raw_data\breast_cancer.csv")
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

- The dataset contains 569 samples.
- 30 numerical features.
- 1 target column.

```python
df.isnull().sum()
```

No missing values are present in the dataset.

```python
df["target"].value_counts()
```

```python
sns.countplot(x="target", data=df)
plt.title("Target Class Distribution")
plt.show()
```

The dataset is relatively balanced between benign and malignant classes.

```python
df.describe()
```

Feature ranges vary significantly.
Some features range from 0–1 while others exceed 1000.
This justifies feature scaling before applying PCA, SVM, or KNN.

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

Strong correlations observed among:
- mean radius, mean perimeter, mean area
- worst radius, worst perimeter, worst area

This indicates multicollinearity and justifies dimensionality reduction.

```python
corr_matrix = df.corr().abs()

upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

high_corr = [
    column for column in upper.columns 
    if any(upper[column] > 0.9)
]

print("Highly correlated features:", high_corr)
```

```python
features = ["mean radius", "mean area", "mean concavity"]

for feature in features:
    sns.histplot(data=df, x=feature, hue="target", kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df.drop("target", axis=1)
y = df["target"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="coolwarm", alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("2D PCA Projection")
plt.colorbar()
plt.show()
```

PCA shows clear separation between benign and malignant samples.
This suggests dimensionality reduction may preserve classification power.

## Key Observations

1. No missing values present.
2. Dataset is reasonably balanced.
3. Significant multicollinearity exists.
4. Feature scales vary widely.
5. PCA visualization shows strong separability.

Conclusion:
Dimensionality reduction techniques are justified 
for reducing redundancy and improving computational efficiency.
